```python
df_pl.head()
```

```python
df_pl["Duels won_p90"] = (df_pl["Duels won, %"] / df_pl["Minutes played"]) * 90
```

```python
df_pl["SDA_p90_pct"] = (
    df_pl["Successful defensive actions per 90"]
    .rank(pct=True) * 100
)

df_pl["Prog_runs_p90_pct"] = (
    df_pl["Progressive runs per 90"]
    .rank(pct=True) * 100
)

df_pl["Key_passes_p90_pct"] = (
    df_pl["Key passes per 90"]
    .rank(pct=True) * 100
)

df_pl["Duels_won_p90_pct"] = (
    df_pl["Duels won_p90"]
    .rank(pct=True) * 100
)

df_pl["Acc_fwd_pass_p90_pct"] = (
    df_pl["Accurate forward passes per 90"]
    .rank(pct=True) * 100
)

df_pl["Acc_prog_pass_p90_pct"] = (
    df_pl["Accurate progressive passes per 90"]
    .rank(pct=True) * 100
)

df_pl["minutes_pct"] = (
    df_pl["Minutes played"]
    .rank(pct=True) * 100
)
```

```python
df_pl.head()
```

```python
df_pl=df_pl.drop(["Minutes played","Team within selected timeframe","Age","Successful defensive actions per 90","Progressive runs per 90","Accurate forward passes, %",
                  "Key passes per 90","Duels won, %","Accurate forward passes per 90","Accurate progressive passes per 90","Duels won_p90"],axis=1)
```

```python
baker = PyPizza(
    params=params,                  # list of parameters
    straight_line_color="#000000",  # color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_lw=1,               # linewidth of last circle
    other_circle_lw=1,              # linewidth for other circles
    other_circle_ls="-."            # linestyle for other circles
)
```

```python
# plot pizza
fig, ax = baker.make_pizza(
    values,              # list of values
    figsize=(7, 9),      # adjust figsize according to your need
    param_location=110,  # where the parameters will be added
    kwargs_slices=dict(
        facecolor="cornflowerblue", edgecolor="#000000",
        zorder=2, linewidth=1
    ),                   # values to be used when plotting slices
    kwargs_params=dict(
        color="#000000", fontsize=12,
        va="center"
    ),                   # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=12,
        zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="cornflowerblue",
            boxstyle="round,pad=0.2", lw=1
        )
    )                    # values to be used when adding parameter-values
)


fig.text(.275,.87,'Casemiro - Manchester United (MUN)',size=18)
fig.text(.275,.85,'per 90 Percentile Rank vs Premier League Midfielders | Season 2025-26', size=12)

team_logo = mpimg.imread("C:/Users/K Raghunandan/Analyst/Logos/Premier League/Manchester United.png")
player_logo = mpimg.imread("C:/Users/K Raghunandan/Analyst/Logos/Players/Casemiro.png")

def add_logo(ax, img, zoom, xy):
    imagebox = OffsetImage(img, zoom=zoom)
    ab = AnnotationBbox(imagebox, xy, frameon=False, xycoords='axes fraction')
    ax.add_artist(ab)

add_logo(ax, team_logo, zoom=0.2, xy=(0.09, 1.1))
add_logo(ax, player_logo, zoom=0.4, xy=(0.06, 1.11))

plt.text(
    0.5, -0.099, "Data Source: DataMB | Viz: @Kandula Raghunandan",
    ha="center", va="center",
    transform=plt.gca().transAxes,
    fontsize=10, color="gray"
)

plt.savefig("C:/Users/K Raghunandan/Analyst/Results/Casemiro_per_90_prem.png",bbox_inches='tight')
```
