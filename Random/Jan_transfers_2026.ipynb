```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import matplotlib.image as mpimg
```

```python
#semenyo
```

```python
df_semenyo = pd.read_excel("C:/Users/K Raghunandan/OneDrive/Desktop/Raghu/Kickoff/CSV Files/semenyo.xlsx")
```

```python
df_semenyo
```

```python
df_semenyo.to_csv("output_file.csv", index=False)
```

```python
df_semenyo
```

```python
df_semenyo=df_semenyo.drop(["number","season","apps","NPxG90xA90"],axis=1)
```

```python
df_semenyo
```

```python
df_semenyo.columns
```

```python
df_semenyo.columns=["team","Goals","Assists","Shots per 90","Passes to Shot per 90","xG per 90","Non-Penalty xG per 90","xA per 90","xG + xA per 90",
                    "xG Chain per 90", "xG Buildup per 90"]
```

```python
df_semenyo
```

```python
params=list(df_semenyo.columns)
params=params[1:]
```

```python
params
```

```python
ranges=[]
bou_values=[]
mci_values=[]


for col in params:
    col_values = pd.to_numeric(df_semenyo[col], errors='coerce')

    a = col_values.min()
    b = col_values.max()

    a = a - (a * 0.25)
    b = b + (b * 0.25)

    ranges.append((a, b))

for x in range(len(df_semenyo["team"])):
    if df_semenyo["team"][x]=="Bournemouth":
        bou_values=df_semenyo.iloc[x].values.tolist()
    if df_semenyo["team"][x]=="Manchester City":
        mci_values=df_semenyo.iloc[x].values.tolist()

bou_values=bou_values[1:]
mci_values=mci_values[1:]
values=[bou_values,mci_values]
```

```python
ranges
```

```python
values
```

```python
radar=Radar()
fig,ax=radar.plot_radar(ranges=ranges,params=params,values=values,radar_color=["#DA020E","#6CABDD"],alphas=[.75,.6],compare=True)

fig.text(0.5, 0.94,
         "Antoine Semenyo",
         ha="center",
         fontsize=18,
         fontweight="bold")

fig.text(0.5, 0.922,
         "For AFC Bournemouth vs For Manchester City",
         ha="center",
         fontsize=12,
         color="black")

fig.text(0.5, 0.905,
         "Premier League per 90' Stats | 202526 Season (So Far MW26)",
         ha="center",
         fontsize=12,
         color="gray")

# Left team
fig.text(0.665, 0.8,
         "AFC Bournemouth",
         ha="center",
         fontsize=14,
         fontweight="bold",
         color="#DA020E")

# Right team
fig.text(0.66, 0.775,
         "Manchester City",
         ha="center",
         fontsize=14,
         fontweight="bold",
         color="#6CABDD")

player_logo = mpimg.imread("C:/Users/K Raghunandan/Analyst/Logos/Players/Semenyo.png")
from_logo = mpimg.imread("C:/Users/K Raghunandan/Analyst/Logos/Premier League/Bournemouth.png")
to_logo = mpimg.imread("C:/Users/K Raghunandan/Analyst/Logos/Premier League/Manchester City.png")
league_logo = mpimg.imread("C:/Users/K Raghunandan/Analyst/Logos/Premier League/PREM LOGO (2).png")

def add_logo(ax, img, zoom, xy):
    imagebox = OffsetImage(img, zoom=zoom)
    ab = AnnotationBbox(imagebox, xy, frameon=False, xycoords='axes fraction')
    ax.add_artist(ab)

add_logo(ax, player_logo, zoom=0.5, xy=(0.45, 1.2))
add_logo(ax, from_logo, zoom=0.2, xy=(0, 1.07))
add_logo(ax, to_logo, zoom=0.2, xy=(0.98, 1.08))
add_logo(ax,league_logo,zoom=0.03, xy=(0.5,0.99))

plt.text(
    0.5, -0.05, "Data Source: Understat | Viz: @Kandula Raghunandan",
    ha="center", va="center",
    transform=plt.gca().transAxes,
    fontsize=10, color="gray"
)

fig.savefig("C:/Users/K Raghunandan/Analyst/Results/semenyo_jan_trans.png", bbox_inches='tight')
```
