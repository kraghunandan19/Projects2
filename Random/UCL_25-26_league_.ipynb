```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import matplotlib.image as mpimg
```

```python
import pandas as pd

df = pd.read_csv("C:/Users/K Raghunandan/Analyst/Project/Data/UCL_League_2025-26.csv")

def clean_team(name):
    return name.split(". ", 1)[-1].strip()

# Clean team names
df['Team_clean_xG'] = df['Team'].apply(clean_team)
df['Team_clean_xGA'] = df['Team.1'].apply(clean_team)

# Separate tables
xg_df = df[['Team_clean_xG', 'xG']].rename(columns={'Team_clean_xG': 'Team'})
xga_df = df[['Team_clean_xGA', 'xGA']].rename(columns={'Team_clean_xGA': 'Team'})

# Merge
df = pd.merge(xg_df, xga_df, on='Team', how='inner')
```

```python
df['xG'] = pd.to_numeric(df['xG'])
df['xGA'] = pd.to_numeric(df['xGA'])
```

```python
plt.rcParams.update({'font.family':'Avenir'})
bgcol = '#fafafa'

fig, ax = plt.subplots(figsize=(10, 8), dpi=120)
fig.set_facecolor(bgcol)
ax.set_facecolor(bgcol)
ax.scatter(df['xG'], df['xGA'], c=bgcol)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_color('#ccc8c8')
ax.spines['bottom'].set_color('#ccc8c8')

plt.tick_params(axis='x', labelsize=12, color='#ccc8c8')
plt.tick_params(axis='y', labelsize=12, color='#ccc8c8')

def getImage(path):
    return OffsetImage(plt.imread(path), zoom=.1, alpha = 1)

for index, row in df.iterrows():
    ab = AnnotationBbox(getImage(row['path']), (row['xG'], row['xGA']), frameon=False)
    ax.add_artist(ab)

plt.hlines(df['xGA'].mean(), df['xG'].min(), df['xG'].max(), color='#c2c1c0')
plt.vlines(df['xG'].mean(), df['xGA'].min(), df['xGA'].max(), color='#c2c1c0')

fig.text(.25,.95,'Champions League 2025-26',size=25)
fig.text(.25,.92,'xG vs xG Against, League Phase', size=15)

league_logo = mpimg.imread("C:/Users/K Raghunandan/Analyst/Logos/UCL 25-26/UCL Logo1.png")

def add_logo(ax, img, zoom, xy):
    imagebox = OffsetImage(img, zoom=zoom)
    ab = AnnotationBbox(imagebox, xy, frameon=False, xycoords='axes fraction')
    ax.add_artist(ab)
    
add_logo(ax, league_logo, zoom=.04, xy=(0.08, 1.1))

fig.text(.06,.14,'xG Against', size=15, color='#575654',rotation=90)
fig.text(.12,0.05,'xG For', size=15, color='#575654')

fig.text(.75,.42,'Avg. xG Against', size=10, color='#575654')
fig.text(.454,.725,'Avg. xG For', size=10, color='#575654',rotation=90)

fig.text(.3,.8,"Low xG, High xGA",color="red",size="8")
fig.text(.75,.34,"High xG, Low xGA",color="red",size="8")

plt.text(
    0.5, -0.099, "Data Source: WhoScored | Viz: @Kandula Raghunandan",
    ha="center", va="center",
    transform=plt.gca().transAxes,
    fontsize=10, color="gray"
)

plt.savefig("C:/Users/K Raghunandan/Analyst/Results/UCL_league_xgvsxga.png",bbox_inches='tight')
```
