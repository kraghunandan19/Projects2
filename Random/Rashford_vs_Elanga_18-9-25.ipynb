```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

```python
df1=pd.read_html("https://fbref.com/en/players/a1d5bd30/all_comps/Marcus-Rashford-Stats---All-Competitions",
                attrs={"id":"last_5_matchlogs"})[0]
```

```python
df2=pd.read_html("https://fbref.com/en/players/2fba6108/all_comps/Anthony-Elanga-Stats---All-Competitions",
                attrs={"id":"last_5_matchlogs"})[0]
```

```python
df=pd.concat([df1,df2]).reset_index()
```

```python
df
```

```python
df.columns
```

```python
df=df.drop(["Date","Day","Comp","Round","Venue","Result","Opponent","Start","Pos","PK","PKatt","CrdR",
            "npxG","Cmp","Att","Carries","Att","Match Report","Name"],axis=1)
```

```python
df
```

```python
df=df.drop(["CrdY","Int","GCA","Succ","index"],axis=1)
```

```python
df
```

```python
df.columns=["Squad","Minutes Played","Goals","Assists","Shots","Shots On Target","Touches","Tackles","Blocks","xG",
            "xAG","Shot Creating Actions","Pass Completion %","Progressive Passes","Progressive Carries"]
```

```python
params=list(df.columns)
params=params[1:]
```

```python
params
```

```python
ranges=[]
a_values=[]
b_values=[]

for x in params:
    a=min(df[params][x])
    a=a-(a*.25)
    
    b=max(df[params][x])
    b=b+(b*.25)

    ranges.append((a,b))

for x in range(len(df["Squad"])):
    if df["Squad"][x]=="es Barcelona":
        a_values=df.iloc[x].values.tolist()
    if df["Squad"][x]=="eng Newcastle Utd":
        b_values=df.iloc[x].values.tolist()

a_values=a_values[1:]
b_values=b_values[1:]
values=[a_values,b_values]
```

```python
ranges
```

```python
values
```

```python
title=dict(
    title_name="Marcus Rashford",
    title_color="#E75480",
    subtitle_name="Against Newcastle 18-09-2025",
    subtitle_color="#E75480",
    title_name_2="Anthony Elanga",
    title_color_2="#241F20",
    subtitle_name_2="Against Barcelona 18-09-2025",
    subtitle_color_2="#241F20",
    title_fontsize=18,
    subtitle_fontsize=15,
)
```

```python
radar=Radar()
fig,ax=radar.plot_radar(ranges=ranges,params=params,values=values,radar_color=["#E75480","#241F20"],alphas=[.75,.6],title=title,compare=True)
plt.text(
    0.5, -0.05, "Data Source: FBRef | Viz: @Kandula Raghunandan",
    ha="center", va="center",
    transform=plt.gca().transAxes,
    fontsize=10, color="gray"
)
plt.savefig("C:/Users/K Raghunandan/Analyst/Results/rashford vs elanga Radar.png")
```
