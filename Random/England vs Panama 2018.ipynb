```python
events=sb.events(match_id=7554)
```

```python
events.head()
```

```python
events.columns
```

```python
events=events[["team","type","minute","location","player","shot_outcome","shot_technique","shot_type","shot_statsbomb_xg","shot_end_location"]]
events=events[(events["team"]=="England") & (events["type"]=="Shot")].reset_index()
```

```python
events.head(20)
```

```python
events[["x","y"]]=pd.DataFrame(events.location.tolist(),index=events.index)
events[["x_end","y_end","height"]]=pd.DataFrame(events.shot_end_location.tolist(),index=events.index)
events["y_new"] = 120 - events["x"]
events["x_new"] = 80 - events["y"]
events["y_end_new"] = 120 - events["x_end"]
events["x_end_new"] = 80 - events["y_end"]
```

```python
fig, ax=plt.subplots(figsize=(10.5,7))
fig.set_facecolor("#22312b")
ax.patch.set_facecolor("#22312b")

pitch = VerticalPitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc',half=True)
pitch.draw(ax=ax, constrained_layout=True, tight_layout=False )
plt.gca().invert_yaxis()

ax.set_xlim(0,80)
ax.set_ylim(0,45)

for x in range(len(events["x"])):
    if events["shot_outcome"][x]=="Saved":
        plt.scatter(events["x_new"][x],events["y_new"][x],color="orange",s=events['shot_statsbomb_xg'][x]*800 + 100,alpha=.7)
    if events["shot_outcome"][x]=="Off T":
        plt.scatter(events["x_new"][x],events["y_new"][x],color="red",s=events['shot_statsbomb_xg'][x]*800 + 100,alpha=.7)
    if events["shot_outcome"][x]=="Blocked":
        plt.scatter(events["x_new"][x],events["y_new"][x],color="blue",s=events['shot_statsbomb_xg'][x]*800 + 100,alpha=.7)
    if events["shot_outcome"][x]=="Goal":
        plt.scatter(events["x_new"][x],events["y_new"][x],color="green",s=events['shot_statsbomb_xg'][x]*800 + 100,alpha=.7)

plt.title("England vs Panama WC 2018 Shotmap",color="white",size="20")
ax.text(2,40,"Goal",color="green")
ax.text(2,38,"Off Target",color="red")
ax.text(2,36,"Saved",color="orange")
ax.text(2,34,"Blocked",color="blue")
```
