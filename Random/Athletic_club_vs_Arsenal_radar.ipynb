```python
df=df.drop(["# Pl","Starts","Min","90s","G-PK","PK","PKatt","CrdR","G-PK_p90","G+A-PK_p90","npxG+xAG","npxG_p90","npxG+xAG_p90","PrgC"],axis=1)
```

```python
df.columns
```

```python
df.columns=["Squad","Avg. Age","Avg. Possesion","Matches Played","Goals","Assists","Goals+Assists","Yellow Cards","xG","Non-Penalty xG","xAG","Progressive Passes",
        "Goals per 90","Assists per 90","Goals + Assists per 90","xG per 90","xAG per 90","xG+xAG per 90","Position"]
```

```python
df.columns
```

```python
params=list(df.columns)
params=params[1:]
```

```python
params
```

```python
ranges=[]
a_values=[]
b_values=[]

for x in params:
    a=min(df[params][x])
    a=a-(a*.25)
    
    b=max(df[params][x])
    b=b+(b*.25)

    ranges.append((a,b))

for x in range(len(df["Squad"])):
    if df["Squad"][x]=="Athletic Club":
        a_values=df.iloc[x].values.tolist()
    if df["Squad"][x]=="Arsenal":
        b_values=df.iloc[x].values.tolist()

a_values=a_values[1:]
b_values=b_values[1:]
values=[a_values,b_values]
```

```python
ranges
```

```python
values
```

```python
title=dict(
    title_name="Athletic Club",
    title_color="#ee2523",
    subtitle_name="2025-26 So far",
    subtitle_color="#ee2523",
    title_name_2="Arsenal",
    title_color_2="#063672",
    subtitle_name_2="2025-26 So far",
    subtitle_color_2="#063672",
    title_fontsize=18,
    subtitle_fontsize=15,
)
```

```python
radar=Radar()
fig,ax=radar.plot_radar(ranges=ranges,params=params,values=values,radar_color=["#ee2523","#063672"],alphas=[.75,.6],title=title,compare=True)
plt.text(
    0.5, -0.05, "Data Source: FBRef | Viz: @Kandula Raghunandan",
    ha="center", va="center",
    transform=plt.gca().transAxes,
    fontsize=10, color="gray"
)
plt.savefig("C:/Users/K Raghunandan/Analyst/Results/Athletic Club vs Arsenal Radar.png")
```
